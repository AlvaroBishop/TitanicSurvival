```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
training_set = pd.read_csv('Train_Titanic.csv')
```

```python
training_set.head()
```

```python
training_set.tail(10)
```

 ## Visualize Dataset

```python
survived = training_set[ training_set['Survived'] == 1]
no_survived = training_set[ training_set['Survived'] == 0]
```

```python
print(f"Total = {len(training_set)} \nNumber of passengers who survived = {len(survived)} \nNumber of passengers who did not survived = {len(no_survived)}")
```

```python
print(f"% Survived = {round((len(survived)/len(training_set) * 100), 2)}% \n% Did not Survived = {round((len(no_survived)/len(training_set) * 100), 2)}%")
```

```python
plt.figure(figsize = [6, 12])
plt.subplot(211)
sns.countplot(x = 'Pclass', data = training_set)
plt.subplot(212)
sns.countplot(x = 'Pclass', hue = 'Survived', data = training_set)
plt.show()
```
